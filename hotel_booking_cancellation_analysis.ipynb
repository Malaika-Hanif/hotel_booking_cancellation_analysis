CANCELLATION ANALYSIS

```python
df['is_canceled'].mean()  # overall cancellation rate
```

 Insight:

The hotel experiences a significant percentage of cancellations, meaning a notable portion of expected revenue is lost before check-in.

Business Impact:

High cancellation reduces revenue predictability and affects room allocation planning.

Recommendation:

Introduce stricter cancellation policies during peak season.

Offer discounts for non-refundable bookings.

```python
df.groupby("hotel")["is_repeated_guest"].mean().sort_values(ascending=False)
```

```python
repeated_guest= df.groupby("hotel")["is_repeated_guest"].mean().sort_values(ascending=False)

plt.figure(figsize=(6,3))
sns.barplot(x=repeated_guest.index ,y=repeated_guest.values)
plt.title("Repeated guest by hotel")
plt.xlabel("hotel")
plt.ylabel("guest_repeated")
plt.show()
```

Insight:

Resort hotel type has a higher percentage of repeated guests, indicating stronger customer loyalty.

Business Meaning:

Higher repeated guest ratio means:

More stable revenue

Lower marketing cost

Better customer satisfaction

 Recommendation:

Focus on loyalty programs and personalized offers to increase repeat bookings, especially for the hotel with lower repeat rate.

```python
df.groupby('hotel')['is_canceled'].mean()  # by hotel type
```

```python
import matplotlib.pyplot  as plt
import seaborn as sns
hotel_cancellation = df.groupby('hotel')['is_canceled'].mean()  # by hotel type
plt.figure(figsize=(6,3)) 
sns.barplot(x=hotel_cancellation.index ,y = hotel_cancellation.values)
plt.title("Cancellation rate by hotel")
plt.xlabel("hotel")
plt.ylabel("cancellation_rate")
plt.show()
```

Insight:

City Hotel has a higher cancellation rate compared to Resort Hotel.

Business Reason:

City hotels often attract short-term business travelers who are more likely to cancel.

Recommendation:

Apply stricter deposit rules for City Hotel.

Send reminder emails closer to check-in date.

REVENUE ANALYSIS

```python
df['revenue'].sum()
```

```python
df.groupby("arrival_date_month")["revenue"].sum().sort_values(ascending=True) # revenue per month
```

```python
revenue_per_month= df.groupby("arrival_date_month")["revenue"].sum().sort_values(ascending=True) # revenue per month
plt.figure(figsize=(12,6))
sns.lineplot(x=revenue_per_month.index ,  y=revenue_per_month.values)
plt.title("Revenue per month")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

Insight:

Certain months generate significantly higher revenue (peak season).

Business Meaning:

Demand is seasonal.

 Recommendation:

Increase prices (dynamic pricing) in peak months.

Run marketing campaigns during low revenue months.

```python
df.groupby("hotel")["revenue"].sum() # revenue per hotel
```

```python
revenue_by_hotel= df.groupby("hotel")["revenue"].sum() # revenue per hotel
plt.figure(figsize=(6,3))
sns.barplot( x= revenue_by_hotel.index , y= revenue_by_hotel.values)
plt.title("Revenue by each hotel")
plt.xlabel("hotel")
plt.ylabel("revenue")
plt.show()
```

Insight:

City hotel type contributes majority of revenue (likely City Hotel due to volume).

 Business Meaning:

Revenue contribution is not equally distributed.

Recommendation:

Invest more in marketing for the stronger revenue hotel.

Improve cancellation control in high-revenue property.
